```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generate some data points for cosine function
x = np.linspace(-np.pi, np.pi, 1000)
y = np.cos(x)

# Define the neural network architecture
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

# Compile the model1 with mean squared error loss and Adam optimizer
model1.compile(loss='mse', optimizer='adam')

# Train the model1 on the generated data
history = model1.fit(x, y, epochs=1000, verbose=0)

# Evaluate the model1 on the training data
train_loss = model1.evaluate(x, y, verbose=0)
print(f'Training Loss: {train_loss:.4f}')

# Plot the results
plt.plot(x, y, label='Cosine')
plt.plot(x, model1.predict(x), label='Approximation')
plt.legend()
plt.show()
```
